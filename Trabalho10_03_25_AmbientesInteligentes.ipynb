Agora vamos renomear as colunas para facilitar o entendimento.

```python
filmes.rename(columns={'movieId':'id_filme', 'title':'Titulo', 'genres':'Generos'}, inplace=True)
tags.rename(columns={'userId':'id_usuario', 'movieId':'id_filme', 'tag':'Tag', 'timestamp':'Timestamp'}, inplace=True)
ratings.rename(columns={'userId':'id_usuario', 'movieId':'id_filme', 'rating':'Nota', 'timestamp':'Timestamp'}, inplace=True)
links.rename(columns={'movieId':'id_filme', 'imdbId':'imdbId', 'tmdbId':'tmdbId'}, inplace=True)
```
